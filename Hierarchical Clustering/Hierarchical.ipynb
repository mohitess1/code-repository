# Hierarchical Clustering

```python
from warnings import filterwarnings
filterwarnings ('ignore')
```

#### Read the Dataset

```python
import pandas as pd
df = pd.read_csv('Mall_Customers (1).csv')
df.head()
```

```python
df.columns=['CustId','Gender','Age','Income','SpendScore']
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

#### Cluster similar customers together on basis of income and Spendscore

```python
X = df[['Income','SpendScore']]
X.head()
```

#### Preprocess the X data

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('imputer',SimpleImputer(strategy='median')),
                           ('scaler',StandardScaler())])
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre
```

#### Apply Hirarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=3)
model.fit(X_pre)
```

```python
lb = model.labels_
```

#### Getting Silhoutte Score for above clustering

```python
from sklearn.metrics import silhouette_score
silhouette_score(X_pre,lb)
```

#### Check for different no of clusters

```python
list(range(2,11))
```

```python
from sklearn.metrics import silhouette_score
scores = []
for i in range(2,11):
    agc = AgglomerativeClustering(n_clusters=i)
    agc.fit(X_pre)
    lb = agc.labels_
    s = silhouette_score(X_pre,lb)
    scores.append(s)

```

```python
scores
```

```python
import matplotlib.pyplot as plt
plt.plot(list(range(2,11)),scores)
plt.xlabel('Number of clusters')
plt.ylabel('silhouette_score')
plt.title('silhouette_scorefor different number of clusters')
plt.show()
```

#### Create a final model which having highest Silhoutte Score

```python
best_agc = AgglomerativeClustering(n_clusters=5)
best_agc.fit(X_pre)
```

```python
final_lb = best_agc.labels_
final_lb
```

#### Add this labesl to X

```python
X['Label'] = final_lb
X
```

```python
for i in range(5):
    print(f'Customer labels : {i}\n')
    display(X[X['Label']==i])
    print('\n#################\n')
```

#### Visualise the clusters

```python
import seaborn as sns
sns.scatterplot(data=X,x='Income',y='SpendScore',hue='Label',palette='deep')
```

#### DendoGram Visualisation

```python
from scipy.spatial import distance_matrix
DM = distance_matrix(X_pre, X_pre)
from scipy.cluster.hierarchy import linkage
linkages = linkage(DM)
from scipy.cluster.hierarchy import dendrogram
plt.figure(figsize=(16,8))
plt.title('Dendogram')
Q = dendrogram(linkages)
```
